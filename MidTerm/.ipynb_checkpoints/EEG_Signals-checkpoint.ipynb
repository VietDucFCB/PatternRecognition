```python
import numpy as np
import scipy.io

class MatlabData:

    def __init__(self, data):
        # Extract data fields
        self.id = data["o"]["id"][0][0][0]
        self.tag = data["o"]["tag"][0][0]
        self.nS = data["o"]["nS"][0][0][0]
        self.sampFreq = data["o"]["sampFreq"][0][0][0][0]
        self.marker = pd.DataFrame(data["o"]["marker"][0][0])
        self.timestamp = data["o"]["timestamp"][0][0]
        self.data =(pd.DataFrame(data["o"]["data"][0,0]))
        self.trials = data["o"]["trials"][0][0]
        self.mapped_data = self.mapping_data(self.data)



    def __repr__(self):
        return (f"MatlabData(\n"
                f"  ID: {self.id},\n"
                f"  Tag:  {self.tag},\n"
                f"  Number of Samples: {pd.DataFrame(self.nS)},\n"
                f"  Sampling Frequency: {self.sampFreq},\n"
                f"  Marker shape: {self.marker},\n"
                f"  Timestamp shape: {self.timestamp.shape},\n"
                f"  Data_mapped : {pd.DataFrame(self.mapped_data)},\n"
                f"  Trials shape (batch size, channels ,S·ªë l∆∞·ª£ng trials,time steps ho·∫∑c samples)  : {self.trials.shape}\n"
                f")")
    @staticmethod
    def mapping_data(dataframe):
      new_names = ['ED_COUNTER' ,'ED_INTERPOLATED', 'ED_RAW_CQ', 'ED_AF3', 'ED_F7', 'ED_F3', 'ED_FC5', 'ED_T7', 'ED_P7', 'ED_O1', 'ED_O2', 'ED_P8','ED_T8', 'ED_FC6','ED_F4','ED_F8','ED_AF4','ED_GYROX','ED_GYROY','ED_TIMESTAMP','ED_ES_TIMESTAMP','ED_FUNC_ID','ED_FUNC_VALUE','ED_MARKER','ED_SYNC_SIGNAL']
      if len(dataframe.columns) != len(new_names):
          raise ValueError("S·ªë l∆∞·ª£ng c·ªôt trong DataFrame v√† s·ªë l∆∞·ª£ng t√™n m·ªõi kh√¥ng kh·ªõp.")
      dataframe.columns = new_names
      return dataframe







mat_file = MatlabData(mat)
print(mat_file.timestamp.shape[0] == mat_file.nS )
```

```python
import numpy as np
import pandas as pd
from scipy.fft import fft, fftfreq
import matplotlib.pyplot as plt

# L·∫•y d·ªØ li·ªáu
time_stamps = pd.to_datetime(mat_datas[0].data['ED_ES_TIMESTAMP'])  # Chuy·ªÉn ƒë·ªïi sang datetime
data_ED_AF3 = mat_datas[0].data['ED_AF3']

# Chuy·ªÉn timestamp v·ªÅ ƒë∆°n v·ªã gi√¢y t·ª´ Unix epoch
time_stamps_seconds = (time_stamps - pd.Timestamp("1970-01-01")).dt.total_seconds()

# T√≠nh kho·∫£ng th·ªùi gian m·∫´u
time_diffs = np.diff(time_stamps_seconds)  # Ch√™nh l·ªách th·ªùi gian gi·ªØa c√°c m·∫´u
if np.allclose(time_diffs, time_diffs[0]):  # Ki·ªÉm tra n·∫øu ƒë·ªÅu ƒë·∫∑n
    sampling_frequency = 1 / time_diffs[0]  # T·∫ßn s·ªë l·∫•y m·∫´u (Hz)
else:
    raise ValueError("Timestamps are not evenly spaced. Resampling is required.")

# Th·ª±c hi·ªán FFT
n_samples = len(data_ED_AF3)
# Convert the Pandas Series to a NumPy array
fft_values = fft(data_ED_AF3.to_numpy())  # FFT t√≠n hi·ªáu
fft_freqs = fftfreq(n_samples, d=1/sampling_frequency)  # T·∫ßn s·ªë t∆∞∆°ng ·ª©ng

# L·∫•y ch·ªâ c√°c t·∫ßn s·ªë d∆∞∆°ng
positive_freqs = fft_freqs[:n_samples // 2]
positive_fft_values = abs(fft_values[:n_samples // 2])  # ƒê·ªô l·ªõn ph·ªï

# V·∫Ω ƒë·ªì th·ªã ph·ªï t·∫ßn s·ªë
plt.figure(figsize=(10, 6))
plt.plot(positive_freqs, positive_fft_values, label="ED_AF3", color='blue')
plt.title("Frequency Spectrum of ED_AF3")
plt.xlabel("Frequency (Hz)")
plt.ylabel("Amplitude")
plt.grid(True)
plt.legend(loc="upper right")
plt.tight_layout()
plt.show()
```
